# Visualize the hourly carbon intensity across regions

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import json
from matplotlib.ticker import MultipleLocator
```

```python
regions = ["US-CAL-CISO", "US-TEX-ERCO", "GB", "AU-SA", "NL"]
fig, axs = plt.subplots(nrows=len(regions), ncols=1, gridspec_kw={'hspace': 0.2, 'wspace': 0.25, 'bottom': 0.2, 
                    'top': 0.8, 'right':0.995, 'left':0.17}, figsize=(6,2.5*len(regions)), sharex=True, sharey=True)

interval = 14*24
starting = 800
for i, region in enumerate(regions):
    df = pd.read_csv(f"{region}_2023_hourly.csv")
    data = df["Carbon Intensity gCOeq/kWh (direct)"]
    axs[i].plot(data[starting:starting+interval], color="black", linewidth=0.5)
    axs[i].set_ylim(0, 450)
```
